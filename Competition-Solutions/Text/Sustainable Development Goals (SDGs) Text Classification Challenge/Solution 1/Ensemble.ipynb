## Ensemble Models
<ul>
<li>Why Ensemble? A combination of non correlated models have been proven to perform better than a single model</li>
<li> I used six models, they were imported below with the score on the leaderboard</li>
<li> The best scoring model scored 0.0411 and the worst scored 0.0480</li>
<li><a href="https://mlwave.com/kaggle-ensembling-guide/"> Ensembling guide</a>
<li> I used sklearn voting classifier to try different weights to apply to the model when ensembling</li>
<li> The final weights were adjusted based on the score of the ensemble on the leaderboard to give a final score of 0.003852</li>

</ul>

```python
#import the libraries
import pandas as pd
import numpy as np
```

```python
#import the class probabilities csv file we created
sub1 = pd.read_csv("submissionlgbp.csv")   #kfold lgb                    0.0411
sub2 = pd.read_csv("submissionbgdtp.csv")  #kfold BaggedDecisiontree    0.0416
sub3 = pd.read_csv("submissiondtp.csv")  #kfold Decisiontree             0.0480
sub4 = pd.read_csv("submissionmlpp.csv")  #kfold mlp                     0.0422
sub5 = pd.read_csv("submissionadp.csv")  #kfold Adaboost                 0.0453
sub6 = pd.read_csv("submissionlogp.csv")  #logisticregression            0.0467
```

```python
#check the correlation between the models for each model
labels = sub1.columns[1:].tolist()
for label in labels:
    print(label)
    print(np.corrcoef([sub1[label].rank(pct=True), sub2[label].rank(pct=True), sub3[label].rank(pct=True), 
                       sub4[label].rank(pct=True), sub5[label].rank(pct=True),
                      sub6[label].rank(pct=True)]))
```

```python
# ensemble the models
average = sub1.copy()
average.iloc[:,1:] = (sub1.iloc[:,1:] * 0.2 + sub2.iloc[:,1:] * 0.3 + sub3.iloc[:,1:] * 0.19
                     +  sub4.iloc[:,1:] * 0.21 + sub5.iloc[:,1:] * 0.1 + sub6.iloc[:,1:] * 0.16)
```

```python
ave = np.around(average.iloc[0:]).astype('int')
```

```python
ave.to_csv('average.csv', index=False)
```
