```python
import pandas as pd
```

```python
one = pd.read_csv("Efficient network B5.csv")
four = pd.read_csv("5 more densenet 201 baseline mean.csv")
five = pd.read_csv("pseudo_labelling - 5 more.csv")
six = pd.read_csv("Hug ME2.csv")
ten = pd.read_csv("Hug ME- 20 epochs of swahili keywords.csv")
eight = pd.read_csv("Hug ME- 20 epochs of swahili.csv")
```

```python
sub = one.copy()
```

```python
for i in ['hapana',
 'kumi',
 'mbili',
 'moja',
 'nane',
 'ndio',
 'nne',
 'saba',
 'sita',
 'tano',
 'tatu',
 'tisa']:
    sub[i] = (one[i]*four[i]*five[i]*six[i]*ten[i]*eight[i])**(1/6)
```
